```python
from keras.datasets import boston_housing
```

```python
(x_train,y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
print((x_train.shape, y_train.shape), (x_test.shape, y_test.shape))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(x_train)
```

```python
x_train = min_max_scaler.fit_transform(x_train)
x_test = min_max_scaler.fit_transform(x_test)
```

```python
x_train[0].shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential(name = 'ANN_Boston_Housing_Price_Prediction')
model.add(Dense(200, input_dim = x_train.shape[1], activation = 'relu', name = 'Input_layer'))
model.add(Dense(150, activation='relu', name = 'Hidden_layer_1'))
model.add(Dense(100, activation= 'relu', name = 'Hidden_layer_2'))
model.add(Dense(50, activation= 'relu', name = 'Hidden_layer_3'))
model.add(Dense(1, activation='linear', name = 'Output_layer'))
```

```python
model.compile(optimizer='Adam', loss = 'mse', metrics = ['mae'])
```

```python
model.summary()
```

```python
hist = model.fit(x_train, y_train, batch_size=40, epochs=500,validation_split= 0.2)
```

```python
hist.history.keys()
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
plt.plot(hist.history['mae'])
plt.plot(hist.history['val_mae'])
```
