# Violin Plot : Similar to Box plot
1. a hybrid of a box plot and a kernel density plot, which shows peaks in the data.

```python
sns.violinplot(x =df['Species'], y =  df['Width'], data = df)
plt.title('Violin plot')
plt.show()
```


# HeatMap : Gives a matrix of all numrical features to each other 
1. tell me the correlation of each column(feature) with each other and show their correlation value as the element of that matrix

```python
Matrix = D.corr()
```

```python
Matrix.shape
```

```python
Matrix
```

```python
sns.heatmap(Matrix, annot = True, cmap = 'coolwarm', linecolor = 'yellow', linewidths = 2)
plt.show()
```

# Pair Plot : visualization of that Matrix that we have made in Heatmap as graphs
1. If both columns are same then it will plot the histograms and other as scatter plots

```python
sns.pairplot(df, hue = 'Species')
```

```python
sns.pairplot(df, hue = 'Species', kind = 'hist')
```

# Dist plot : Visualization of single feature(numerical) 
1. Show me the gaussian curve and histogram of that feature value 

```python
sns.distplot(df['Height'], label = 'Gaussian distribution of Height')
# sns.distplot?
```

```python
sns.distplot(df['Height'], hist = False)
```

# Kde plot : Same as Dist plot but it will only show me the Gaussian curve of that feature value

# Regression plot : Regression(prediction curve) line of two numerical feature value

```python
sns.regplot(x = 'Weight', y = 'Length1', data = df)
```

# Lm Plot : Similar to Regression plot but here we can visualize this prediction line for each category present in that categorical Feature

```python
sns.lmplot(x = 'Weight', y = 'Length1', data = df, col = 'Species')
```
