# Histogram : 
1. Visualization of gaussian of any numerical column present in dataset
1. Visualization of any Gaussian Distribution 



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_gaussian = np.random.randn(1000)
sigma = 5
mean = 60

x_shifted_gausian_1 = x_gaussian*sigma + mean

x_shifted_gausian_2 = x_gaussian*7 + 70
```

```python
x_shifted_gausian_1 = x_shifted_gausian_1.round()
x_shifted_gausian_2 = x_shifted_gausian_2.round()
```

```python
plt.hist(x_shifted_gausian_1, bins = 20)
plt.hist(x_shifted_gausian_2, alpha = 0.5, bins = 15)

# sns.distplot(x_shifted_gausian_2, hist = False)
```
