```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(20)
```

```python
x
```

```python
y1 = 3*x + 7
```

```python
y2 = 4*x**2 + 6
```

#  Line plot : For Continous Data

1. Categorical Column with any Other Numerical Column to check how the specific category varies for that Column
2. Relationship between two Numerical Columns

```python
plt.plot(x, y1, color = 'red', linestyle = '--', label = 'line plot')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Velocity time graph')
plt.legend()
plt.show()
```

```python
plt.plot(x, y2, color = 'cyan', linestyle = '--', label = 'line plot')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Velocity time graph')
plt.legend()
plt.show()
```

#  Scatter plot : For Discrete Data

1. Categorical Column with any Other Numerical Column to check how the specific category varies for that Column
2. Relationship between two Numerical Columns

```python
plt.scatter(x , y1 , c = 'green', label = 'Scatter plot' , marker = '^')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Velocity time graph')
plt.legend()
plt.show()
```

```python
plt.scatter(x , y2 , c = 'green', label = 'Scatter plot' , marker = '*')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Velocity time graph')
plt.legend()
plt.show()
```

# Bar Plot : For Categorical Data
Visualization of Range of Any Categorical Data values present in that columns with some other Numerical Columns(parameter)

```python
plt.figure(figsize=(3,5))
plt.bar(['Sidhu', 'Karan Aujla','Arijit Singh'], [20, 15, 10])
plt.show()
```

```python
x_cor = np.array([0,1,2])
x_cor + 0.5
```

```python
plt.bar(x_cor, [10, 15, 20], width = 0.5, label = 'Blue Data')
plt.bar(x_cor + 0.5, [20, 15, 10], width = 0.5 , label = 'Red data', color = 'red' )
plt.ylim(0, 30)
plt.xlim(-1, 5)
plt.ylabel('Sales')
plt.xlabel('Time')
plt.title('Bar Plot')
plt.legend()
plt.show()
```

# Histogram : 
1. Visualization of gaussian of any numerical column present in dataset
1. Visualization of any Gaussian Distribution 



```python
x_gaussian = np.random.randn(1000)
sigma = 5
mean = 60

x_shifted_gausian_1 = x_gaussian*sigma + mean

x_shifted_gausian_2 = x_gaussian*7 + 70
```

```python
x_shifted_gausian_1 = x_shifted_gausian_1.round()
x_shifted_gausian_2 = x_shifted_gausian_2.round()
```

```python
plt.hist(x_shifted_gausian_1)
plt.hist(x_shifted_gausian_2, alpha = 0.5)

# sns.distplot(x_shifted_gausian_2, hist = False)
```
