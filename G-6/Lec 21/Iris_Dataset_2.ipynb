```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data= iris.data
```

```python
data = pd.DataFrame(data, columns = iris.feature_names)
```

```python
data.head()
```

```python
data['target'] = iris.target
```

```python
data['Flower_name'] = iris.target_names[data['target']]
```

```python
data.head()
```

```python
setosa = data[data['Flower_name'] == 'setosa']
versicolor = data[data['Flower_name'] == 'versicolor']
verginica = data[data['Flower_name'] == 'verginica']
```

```python
plt.scatter(data['sepal length (cm)'], data['sepal width (cm)'])
```

```python
plt.scatter(data['petal length (cm)'], data['petal width (cm)'])
```

```python
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], marker = '*', label = 'Setosa', color = 'Red')
plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], marker = '^', label = 'Versicolor', color = 'yellow')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Setosa vs Versicoor')
plt.legend()
plt.show()
```

```python
plt.scatter(verginica['petal length (cm)'], verginica['petal width (cm)'], marker = '*', label = 'Verginica', color = 'Red')
plt.scatter(versicolor['petal length (cm)'], versicolor['petal width (cm)'], marker = '^', label = 'Versicolor', color = 'yellow')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.title('Virginica vs Versicoor')
plt.legend()
plt.show()
```
