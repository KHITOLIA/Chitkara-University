```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
`df = pd.read_csv('Fish.csv')
df.head()
```

# Line Plot : Visualization of Two different features (continous data)
1. Categorical Feature with Numerical Feature 
2. Numerical Feature with Numerical Feature

```python
sns.lineplot(x = df['Species'], y = df['Weight'], data = df)
```

```python
sns.lineplot(x = df['Length1'], y = df['Weight'], data = df)
```

# Scatter Plot : Visualization of Two different features (Discrete Data)
1. Categorical Feature with Numerical Feature 
2. Numerical Feature with Numerical Feature

```python
sns.scatterplot(x =df['Species'], y = df['Length1'], data = df, marker = "^")
```

```python
sns.scatterplot(x =df['Weight'], y = df['Length1'], data = df, marker = "^")
```

# Bar Plot : Bar Graph of Categorcical Feature with one Numerical Feature 
1. How much it will rise for that particular Feature value

```python
sns.barplot(x = df['Species'], y = df['Length1'], data = df)
plt.title('Bar plot')
plt.show()
```

# Count Plot : Same as our Pie chart 
1. Seperately Count the Categories present in that Categorical Feature and visualize it as a Bar Graph

```python
sns.countplot(x  =df['Species'], data = df)
```

# Box Plot : Visualization of one Categorical feature value with different Numerical feature value
1. How does it's range (25% - 75%) varies for that numrical feature value of each cateogry present in that catogorical column

```python
sns.boxplot(x =df['Species'], y =  df['Length1'], data = df)
```

# Violin Plot : Similar to Box plot
1. a hybrid of a box plot and a kernel density plot, which shows peaks in the data.

```python
sns.violinplot(x =df['Species'], y =  df['Width'], data = df)
plt.title('Violin plot')
plt.show()
```

# Dist plot : Visualization of single feature(numerical) 
1. Show me the gaussian curve and histogram of that feature value 

```python
sns.distplot(df['Height'], hist = False)
```

```python
sns.distplot(df['Height'], hist = False)
```

# Regression plot : Regression(prediction curve) line of two numerical feature value

```python
sns.regplot(x = 'Weight', y = 'Length1', data = df)
```

# Lm Plot : Similar to Regression plot but here we can visualize this prediction line for each category present in that categorical Feature

```python
sns.lmplot(x = 'Weight', y = 'Length1', data = df, col = 'Species')
```
